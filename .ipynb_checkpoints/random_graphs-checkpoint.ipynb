```python
import random
import networkx as nx
```

```python
def generate_connected_dag(n, p):
    """
    Generates a simple connected random DAG with n nodes and edge probability p.
    
    One of the nodes is labelled Y, and is the last one in the topological order. 
    
    Parameters:
        n (int): Number of nodes.
        p (float): Probability of additional edge creation.
        
    Returns:
        G (nx.DiGraph): A directed acyclic graph (DAG) that is weakly connected.
    """
    G = nx.DiGraph()
    #G.add_nodes_from(range(n))

    # Ensure the graph is connected by creating a random path connecting all nodes
    nodes = [f"X{i}" for i in range(n-1)]
    random.shuffle(nodes) # Random topological order
    nodes +=  ["Y"]
    for i in range(n - 1):
        G.add_edge(nodes[i], nodes[i + 1])

    # Add additional random edges while keeping the graph acyclic
    for i in range(n):
        for j in range(i + 1, n):
            if random.random() < p:
                G.add_edge(nodes[i], nodes[j])

    return G
```

```python
# Example usage: Generate a connected DAG with 10 nodes and edge probability 0.3
dag = generate_connected_dag(8, 0.3)
nx.draw(dag, with_labels=True)
```
